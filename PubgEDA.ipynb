```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import Normalizer

df.dropna(subset=["winPlacePerc"],inplace=True)

X=df.drop(["Id","groupId","matchId","matchType","winPlacePerc"],axis=1)
y=df["winPlacePerc"]

col_names=X.columns
transformer=Normalizer().fit(X)
X=transformer.transform(X)
```

```python
X=pd.DataFrame(X,columns=col_names)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)

D_train=xgb.DMatrix(X_train,label=y_train)
D_test=xgb.DMatrix(X_test,label=y_test)
```

```python
param={
        'eta':0.15,
        'max_depth':5,
        'num_class':2}

steps=2
model=xgb.train(param,D_train,steps)
```

```python
fig, ax1=plt.subplots(figsize=(8,15))
xgb.plot_importance(model,ax=ax1)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
preds=model.predict(D_test)
best_preds=np.asarray([np.argmax(line) for line in preds])
print("MSE = {}".format(mean_squared_error(y_test,best_preds)))
```
